## **2.3 Converting Data Type**
This code performs data type standardization and value transformation to prepare the dataset for clearer analysis and interpretation.

A list of categorical columns (gender, job_type, social_platform_preference) is defined and converted to the category data type.

## **2.4 Handling Zero Values in Numerical Columns**

**Objective**:
Certain numerical columns contain zero values that do not represent valid or realistic measurements. If left untreated, these values can distort statistical summaries and analytical results.
Before applying this transformation, the dataset included zero values in key productivity, wellbeing, and digital usage columns. These zeros could negatively impact mean calculations, trends, and visual interpretations, leading to biased insights.

**Code Logic Explanation**

* A list of columns requiring correction is defined.

* For each column:

* Zero values (0) are treated as missing data by replacing them with NaN.

* The mean value of the column is calculated while excluding missing values.

* All zero entries are then replaced with the computed mean, preserving data consistency.
This ensures that unrealistic zero values do not skew the dataset.

```python
cols_to_fix =[ "daily_social_media_time",
  "number_of_notifications","work_hours_per_day",
  "perceived_productivity_score" , "actual_productivity_score",
  "stress_level","sleep_hours","screen_time_before_sleep" ,  "uses_focus_apps"  ,
  "has_digital_wellbeing_enabled" , "days_feeling_burnout_per_month",
  "weekly_offline_hours",
  "job_satisfaction_score"]

for col in cols_to_fix:
    mean_value = df[col].replace(0, pd.NA).mean()  # Treat 0 as missing
    df[col] = df[col].replace(0, mean_value)
```

**After running the code:**

* All invalid zero values are replaced with meaningful average values.

* The numerical distributions become more balanced and representative.

* The dataset remains complete, with no missing values.

* Subsequent analysis and visualizations produce more accurate and trustworthy results.

**Final Insight:**
Replacing invalid zero values with column means enhances data quality, reliability, and analytical accuracy, making the dataset well-prepared for exploratory analysis and modeling.

# **3.2.3 Multivariate Analysis**
Multivariate analysis is a statistical approach used to analyze and understand relationships among three or more variables simultaneously. Unlike univariate or bivariate analysis, it captures complex interactions, patterns, and dependencies between multiple factors, making it especially useful for studying real-world behaviors such as the combined impact of social media usage, lifestyle habits, and productivity.

**Multivariate Analysis Using Different Plots**
* Correlation Heatmap â€“ Productivity Correlations
Shows the strength and direction of linear relationships between all selected variables simultaneously. It helps quickly identify strong, weak, or negligible associations, especially between social media usage and productivity scores.

* Pairplot â€“ Productivity Relationships by Gender
Displays pairwise relationships among multiple numerical variables at once. It helps identify distribution patterns, potential relationships, and gender-based variations in social media usage and productivity metrics.

* Clustermap â€“ Clustered Correlation Heatmap of Productivity Variables
Applies hierarchical clustering to group variables with similar correlation patterns. It reveals natural groupings, distinguishing productivity-related variables from lifestyle and social media behaviors.

## **CORRELATION HEATMAP**

* Selects key numerical productivity and lifestyle variables for analysis.

* Computes the Pearson correlation matrix to measure linear relationships.

* Visualizes correlations using a heatmap with annotated values.

* Uses a diverging color scale centered at zero to clearly distinguish positive and negative relationships.

* Supports Exploratory Data Analysis (EDA) by quickly identifying strong and weak associations among variables.

```python
# 8. Correlation Heatmap (Numeric columns)
plt.figure(figsize=(10,6))
numeric_cols = ['daily_social_media_time', 'work_hours_per_day', 'sleep_hours',
                'perceived_productivity_score', 'actual_productivity_score',
                'stress_level', 'screen_time_before_sleep']
corr = df[numeric_cols].corr()
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f", center=0)
plt.title('Productivity Correlations')
plt.show()
```

**Key Insights from the Visualization**

* Perceived productivity score and actual productivity score show a very strong positive correlation (~0.96), indicating high consistency between self-reported and actual productivity.

* Most lifestyle variables (daily social media time, sleep hours, screen time before sleep) show very weak or near-zero correlations with productivity scores.

* Work hours per day has minimal correlation with both perceived and actual productivity, suggesting longer hours do not directly imply higher productivity.

* Stress level shows a slight negative or negligible relationship with productivity, indicating limited linear impact.

* Overall, the heatmap confirms that strong linear relationships are rare, and productivity is influenced by multiple factors beyond simple time-based metrics.

## **PAIR PLOT**
* Creates a pairplot of the first four numerical variables to visualize pairwise relationships.

* Displays scatter plots for every numerical variable combination.

* Shows KDE (density) plots on the diagonal to understand individual distributions.

* Uses gender as a hue to compare patterns across Female, Male, and Other groups.

* Helps in exploratory data analysis (EDA) to detect relationships, trends, and group-wise differences in productivity-related metrics.

```python
# 9. Pairplot (Key numeric vars + gender)
sns.pairplot(df[numeric_cols[:4] + ['gender']], hue='gender', palette='bright',
             diag_kind='kde')
plt.suptitle('Productivity Relationships by Gender', y=1.02)
plt.show()
```

**Key Insights from the Visualization**


* Distributions overlap heavily across genders, indicating no strong gender-based separation in productivity metrics.


* Diagonal KDE plots show similar distribution shapes for all genders, suggesting comparable ranges and central tendencies.


* Scatter plots reveal weak to moderate relationships between variables such as work hours, sleep hours, and productivity scores.


* No clear linear correlation dominates between most variable pairs.


* Perceived productivity score appears widely spread, indicating varied self-assessment across individuals regardless of gender.


* Overall, gender does not strongly influence productivity patterns, and variations are more individual-driven than group-driven.

## **CLUSTERMAP**

* Computes the correlation matrix for key productivity-related numerical variables.

* Uses a clustermap to apply hierarchical clustering on both rows and columns.

* Groups variables based on similar correlation patterns, not just individual values.

* Displays a heatmap with annotations to clearly show correlation strengths.

* Adds dendrograms to visualize how variables are clustered together.

* Supports advanced EDA by revealing hidden structure among productivity metrics.

```python
# 10 Clustermap - Hierarchical Clustering of Productivity Metrics
plt.figure(figsize=(12,10))
numeric_cols = ['daily_social_media_time', 'work_hours_per_day', 'sleep_hours',
                'perceived_productivity_score', 'actual_productivity_score',
                'stress_level', 'screen_time_before_sleep']

# Cluster rows (observations) and columns (variables)
sns.clustermap(df[numeric_cols].corr(),
               annot=True, cmap='coolwarm', fmt=".2f",
               linewidths=0.5, figsize=(12,8),
               dendrogram_ratio=0.1)
plt.title('Clustered Correlation Heatmap: Productivity Variables')
plt.tight_layout()
plt.show()
```

**Key Insights from the Visualization**

* Perceived productivity score and actual productivity score form a strong cluster with a very high positive correlation (~0.96), indicating strong alignment between self-reported and measured productivity.

* Most other variables show very weak or near-zero correlations, suggesting limited linear relationships.

* Daily social media time, screen time before sleep, and sleep hours cluster together, reflecting related lifestyle behaviors.

* Work hours per day and stress level show minimal correlation with productivity scores, implying productivity is not solely driven by longer working hours or stress.

* The clustering confirms that productivity metrics are more closely related to each other than to lifestyle or time-usage variables.

* Pairplot â€“ Productivity Relationships by Gender
Shows interactions among multiple variables with gender as a grouping factor. Patterns largely overlap across genders, indicating no strong gender-based differences in productivity or social media usage.

* Clustermap â€“ Clustered Correlation Heatmap: Productivity Variables
Groups variables based on similarity. Productivity scores cluster together, while social media, sleep, and screen-time variables form a separate lifestyle cluster, indicating indirect influence.

Productivity measures are strongly aligned with each other, whereas social media usage and lifestyle variables show minimal direct impact on productivity in a multivariate context.
